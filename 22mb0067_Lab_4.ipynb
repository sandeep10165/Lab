```python
#Q4. Select tuples from instructor where dept_name = ‘History’
import mysql.connector

cnx = mysql.connector.connect(host ="localhost",
port="3306",
user="root",
password="",
database="database_name")
cursor = cnx.cursor()


query = "SELECT * FROM instructor WHERE dept_name = %s"
dept = ('History',)
cursor.execute(query, dept)

result = cursor.fetchall()
for instructor in result:
    print(instructor)

cursor.close()
cnx.close()
```

```python
#Q6. Find the names of all instructors who have taught some course and the course_id
import mysql.connector

cnx = mysql.connector.connect(host ="localhost",
port="3306",
user="root",
password="",
database="database_name")
cursor = cnx.cursor()

query = "SELECT i.name, t.course_id FROM instructor i, teaches t WHERE i.ID = t.ID"
cursor.execute(query)

result = cursor.fetchall()
for row in result:
    print(row)

cursor.close()
cnx.close()
```

```python
#Q7. Find the names of all instructors whose name includes the substring “dar”.
import mysql.connector

cnx = mysql.connector.connect(host ="localhost",
port="3306",
user="root",
password="",
database="database_name")
cursor = cnx.cursor()

query = "SELECT name FROM instructor WHERE name LIKE %s"
substring = ('%dar%',)
cursor.execute(query, substring)

result = cursor.fetchall()
for row in result:
    print(row)

cursor.close()
cnx.close()
```

```python
# Q8. Find the names of all instructors with salary between 90,000 and 100,000 (that is, ≥ 90,000 and ≤ 100,000)

import mysql.connector

cnx = mysql.connector.connect(host ="localhost",
port="3306",
user="root",
password="",
database="database_name")
cursor = cnx.cursor()

query = "SELECT name FROM instructor WHERE salary BETWEEN %s AND %s"
salary_range = (90000, 100000)
cursor.execute(query, salary_range)

result = cursor.fetchall()
for row in result:
    print(row)

cursor.close()
cnx.close()
```
